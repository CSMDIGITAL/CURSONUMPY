```python
# Carregar dataset
df = pd.read_excel("datasets/petrobras.xlsx")

```

```python
# Garantir nomes de colunas como string
df.columns = [str(c) for c in df.columns]
```

```python
# Selecionar a primeira coluna numérica (pois o nome não é '0')
coluna_preco = df.columns[0]
```

```python
# Extrair todos os preços
precos = df[coluna_preco].dropna().values
```

```python
# Criar eixo X com índices 0,1,2,...,19
indices = list(range(len(precos)))
```

```python
# Criar gráfico completo
plt.plot(indices, precos, marker='o')
plt.title("Variação Completa do Preço - Petrobras")
plt.xlabel("Índice")
plt.ylabel("Preço (R$)")
plt.grid(True)
plt.show()
```
