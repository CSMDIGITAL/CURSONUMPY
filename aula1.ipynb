#### Iris (O nosso Hello Data Cience!!!)
- O conjunto de dados Iris é um clássico em estatística e aprendizado de máquina. 
- Os dados foram coletados por Edgar Anderson e popularizados por Ronald A. Fisher em 1936 no artigo "The use of multiple measurements in taxonomic problems", onde Fisher usou-os para demonstrar análise discriminante linear. 
- Contém 150 amostras de três espécies de Iris (I. setosa, I. versicolor, I. virginica), com quatro medidas: comprimento/largura das sépalas e pétalas. 
- Por ser pequeno, bem comportado e fácil de visualizar, tornou-se um exemplo padrão para ensinar métodos de classificação, visualização e validação de modelos. 
- Hoje é amplamente disponível em repositórios como o UCI Machine Learning Repository e em bibliotecas estatísticas e de ML.

```python
# proponha operações no array
print(f"média de cada coluna{data_iris.mean(axis=0)}")  # média de cada coluna
print(f"desvio padrão de cada coluna {data_iris.std(axis=0)}")  # desvio padrão de cada coluna
print(f"valor máximo de cada coluna {data_iris.max(axis=0)}")  # valor máximo de cada coluna
print(f"valor mínimo de cada coluna {data_iris.min(axis=0)}")  # valor mínimo de cada coluna
# calcular a média da coluna 0
data_iris[:,0].mean()
data_iris[:,0].std()
data_iris[:,0].max()
data_iris[:,0].min()
data_iris[:,0].sum()
data_iris[:,0].var()  # variância da coluna 0
data_iris[:,0].cumsum()  # soma acumulada da coluna 0
data_iris[:,0].prod()  # produto dos valores da coluna 0
data_iris[:,0].round(2)  # arredondar os valores da coluna 0 para 2 casas decimais
data_iris[:,0].astype(int)  # converter os valores da coluna 0 para inteiros
data_iris[:,0] > 5.0  # valores da coluna 0 maiores que 5.0
data_iris[data_iris[:,0] > 5.0]  # linhas onde a coluna 0 é maior que 5.0
data_iris[:,0][data_iris[:,0] > 5.0]  # valores da coluna 0 maiores que 5.0
data_iris[:,0].argsort()  # índices que ordenariam a coluna 0
data_iris[data_iris[:,0].argsort()]  # ordenar o array pela coluna 0
```

```python
# faça o histograma da coluna 1 do dataset Iris
import matplotlib.pyplot as plt
plt.hist(data_iris[:,1], bins=10)
plt.title('Histograma da coluna 1 do dataset Iris')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

Leitura do Histograma

    Eixo X (Horizontal - "Valor"): Representa os valores da coluna 1 do dataset Iris. O eixo está dividido em intervalos, chamados de bins (compartimentos ou classes).

        Exemplo dos bins: Os valores vão de 2.0 a 4.5, com divisões claras em 2.0, 2.5, 3.0, 3.5, 4.0, e 4.5.

    Eixo Y (Vertical - "Frequência"): Representa a contagem ou o número de ocorrências (frequência) de valores que caem dentro de cada intervalo (bin) do eixo X.

🔍 Interpretação dos Dados

Analisando a altura das barras, podemos tirar as seguintes conclusões:

    Maior Frequência (Moda): A barra mais alta está centrada ligeiramente acima do valor 3.0.

        Isso significa que a maior concentração de valores na coluna 1 do dataset Iris está no intervalo entre aproximadamente 2.9 e 3.1 (o bin correspondente a 3.0 tem uma frequência de 38, o valor máximo no Y).

    Segunda Maior Frequência: O segundo pico está no intervalo entre 3.0 e 3.5. A barra centrada em 3.25 tem uma frequência de cerca de 31.

    Distribuição dos Dados:

        A distribuição não é perfeitamente simétrica. Ela parece ter um leve viés (assimetria) para a direita (a cauda estende-se um pouco mais para a direita), mas é mais ou menos em forma de sino (distribuição normal) em sua parte principal.

        Valores Baixos: Os valores mais baixos, no intervalo de 2.0 a 2.25, têm uma frequência baixa (cerca de 4).

        Valores Altos: Os valores mais altos (acima de 4.0) também têm uma frequência muito baixa (cerca de 2).

    Conclusão: A maioria dos dados (a maior parte da coluna 1) está concentrada na faixa de 2.5 a 3.5, com um pico muito forte por volta de 3.0.

Em resumo, o histograma mostra que a característica representada pela coluna 1 no dataset Iris varia principalmente entre 2.0 e 4.5, sendo 3.0 o valor mais comum.
